# 单调函数（CS/AI 专项笔记·进阶版）

## 1. 严格定义（大学阶段纯数学表述）

### 1.1 单调递增函数（严格+非严格）

#### （1）非严格单调递增（单调不减）

- **文字描述**：设函数 $f(x)$ 的定义域为 $D$，区间 $I \subseteq D$，若对**任意** $x_1, x_2 \in I$，当 $x_1 < x_2$ 时，恒有 $f(x_1) \leq f(x_2)$，则称 $f(x)$ 在 $I$ 上**单调递增**。
- **数学描述**：$\forall x_1, x_2 \in I$，$x_1 < x_2 \implies f(x_1) \leq f(x_2)$。

#### （2）严格单调递增

- **文字描述**：设函数 $f(x)$ 的定义域为 $D$，区间 $I \subseteq D$，若对**任意** $x_1, x_2 \in I$，当 $x_1 < x_2$ 时，恒有 $f(x_1) < f(x_2)$，则称 $f(x)$ 在 $I$ 上**严格单调递增**。
- **数学描述**：$\forall x_1, x_2 \in I$，$x_1 < x_2 \implies f(x_1) < f(x_2)$。

### 1.2 单调递减函数（严格+非严格）

#### （1）非严格单调递减（单调不增）

- **文字描述**：设函数 $f(x)$ 的定义域为 $D$，区间 $I \subseteq D$，若对**任意** $x_1, x_2 \in I$，当 $x_1 < x_2$ 时，恒有 $f(x_1) \geq f(x_2)$，则称 $f(x)$ 在 $I$ 上**单调递减**。
- **数学描述**：$\forall x_1, x_2 \in I$，$x_1 < x_2 \implies f(x_1) \geq f(x_2)$。

#### （2）严格单调递减

- **文字描述**：设函数 $f(x)$ 的定义域为 $D$，区间 $I \subseteq D$，若对**任意** $x_1, x_2 \in I$，当 $x_1 < x_2$ 时，恒有 $f(x_1) > f(x_2)$，则称 $f(x)$ 在 $I$ 上**严格单调递减**。
- **数学描述**：$\forall x_1, x_2 \in I$，$x_1 < x_2 \implies f(x_1) > f(x_2)$。

### 1.3 核心限定说明

- **区间依赖性**：单调性是“区间局部性质”，而非定义域全局性质（如 $f(x)=x^2$ 在 $[0,+\infty)$ 严格递增，在 $(-\infty,0]$ 严格递减）；
- **任意性要求**：需对区间内所有成对自变量满足关系，而非部分点（如 $f(x)=\begin{cases}x, & x \neq 1 \\ 2, & x=1\end{cases}$ 在 $[0,2]$ 上不严格递增，因存在 $x_1=1 < x_2=1.5$ 但 $f(x_1)=2 > f(x_2)=1.5$）；
- **严格性意义**：严格单调性是函数“单射”（一一对应）的充分条件，也是反函数存在的核心前提。

## 2. 核心性质与原理

### 2.1 核心要素（单调函数的三大本质特征）

1. **顺序保持性**：严格单调函数保持自变量与函数值的“严格顺序对应”（$x_1 < x_2 \iff f(x_1) < f(x_2)$ 或 $f(x_1) > f(x_2)$）；
2. **区间封闭性**：单调函数在子区间上仍保持原单调性（如 $f(x)$ 在 $[a,b]$ 严格递增，则在 $[a,c]$（$c \in (a,b)$）上也严格递增）；
3. **最值存在性**：闭区间上的单调函数必在端点处取得最值（严格递增函数在左端点取最小值，右端点取最大值；严格递减函数反之）。

### 2.3 本质原理

单调函数的核心是“自变量与函数值的顺序一致性”：

- 严格递增函数：自变量的“小于”关系严格对应函数值的“小于”关系（顺序完全保真）；
- 严格递减函数：自变量的“小于”关系严格对应函数值的“大于”关系（顺序完全反转）；
  这种顺序一致性使得单调函数具备“可预测性”和“可逆性”，是算法设计（如二分查找）、模型优化（如梯度下降）、数据映射（如特征编码）的核心理论基础。

## 3. 关键公式与推导

### 3.1 严格单调性的量化判定公式

$$f(x) \text{ 在 } I \text{ 上严格递增} \iff \forall x_1, x_2 \in I, (x_1 - x_2)(f(x_1) - f(x_2)) > 0$$
$$f(x) \text{ 在 } I \text{ 上严格递减} \iff \forall x_1, x_2 \in I, (x_1 - x_2)(f(x_1) - f(x_2)) < 0$$

**推导逻辑**：

- 若 $x_1 < x_2$，则 $x_1 - x_2 < 0$，此时 $(x_1 - x_2)(f(x_1) - f(x_2)) > 0 \implies f(x_1) - f(x_2) < 0 \implies f(x_1) < f(x_2)$，符合严格递增定义；
- 同理可推导严格递减的等价条件，该公式可直接用于单调性的代数证明。

### 3.2 复合函数单调性推导（以“同增”为例）

已知：$f(u)$ 在 $J$ 上严格递增，$g(x)$ 在 $I$ 上严格递增，求证 $f(g(x))$ 在 $I$ 上严格递增。

**推导过程**：

1. 任取 $x_1 < x_2 \in I$；
2. 因 $g(x)$ 严格递增，故 $g(x_1) < g(x_2)$，且 $g(x_1), g(x_2) \in J$（$g$ 的值域是 $f$ 的定义域子集）；
3. 因 $f(u)$ 严格递增，且 $g(x_1) < g(x_2)$，故 $f(g(x_1)) < f(g(x_2))$；
4. 由 $x_1 < x_2$ 可推出 $f(g(x_1)) < f(g(x_2))$，符合严格递增定义，故 $f(g(x))$ 在 $I$ 上严格递增。


## 4. 例题解析（大学函数章节高频题型）

### 例题 1：代数法证明分式函数单调性

**题目**：证明 $f(x) = \frac{x}{1 + x^2}$ 在 $[1, +\infty)$ 上严格递减。

**证明思路**：

1. 任取 $x_1 < x_2 \in [1, +\infty)$；
2. 计算 $(x_1 - x_2)(f(x_1) - f(x_2))$，证明其小于 0（严格递减等价条件）；
3. 化简表达式，利用区间性质分析符号。

**详细证明过程**：

1. 任取 $x_1 < x_2 \in [1, +\infty)$；

2. 计算 $f(x_1) - f(x_2)$：
   $$
   f(x_1) - f(x_2) = \frac{x_1}{1 + x_1^2} - \frac{x_2}{1 + x_2^2} = \frac{x_1(1 + x_2^2) - x_2(1 + x_1^2)}{(1 + x_1^2)(1 + x_2^2)}
   $$

3. 分子化简：
   $$
   x_1(1 + x_2^2) - x_2(1 + x_1^2) = x_1 - x_2 + x_1x_2^2 - x_2x_1^2 = (x_1 - x_2) - x_1x_2(x_1 - x_2) = (x_1 - x_2)(1 - x_1x_2)
   $$

4. 分析各项符号：

   - $x_1 < x_2 \implies x_1 - x_2 < 0$；
   - $x_1, x_2 \geq 1 \implies x_1x_2 \geq 1 \implies 1 - x_1x_2 \leq 0$（且 $x_1 \neq x_2$，故 $1 - x_1x_2 < 0$）；
   - 分母 $(1 + x_1^2)(1 + x_2^2) > 0$（平方项恒正）；

5. 计算乘积符号：
   $$
   (x_1 - x_2)(f(x_1) - f(x_2)) = (x_1 - x_2) \cdot \frac{(x_1 - x_2)(1 - x_1x_2)}{(1 + x_1^2)(1 + x_2^2)} = \frac{(x_1 - x_2)^2(1 - x_1x_2)}{(1 + x_1^2)(1 + x_2^2)}
   $$
   其中 $(x_1 - x_2)^2 > 0$，$1 - x_1x_2 < 0$，分母 $> 0$，故整体 $< 0$；

6. 结论：$\forall x_1 < x_2 \in [1, +\infty)$，$(x_1 - x_2)(f(x_1) - f(x_2)) < 0$，故 $f(x)$ 在 $[1, +\infty)$ 上严格递减。

### 例题 2：复合函数单调性证明（同增异减）

**题目**：已知 $f(u) = e^u$（$u \in \mathbb{R}$）严格递增，$g(x) = -x^2 + 2x$（$x \in (-\infty, 1]$）严格递增，证明 $h(x) = f(g(x)) = e^{-x^2 + 2x}$ 在 $(-\infty, 1]$ 上严格递增。

**证明过程**：

1. 任取 $x_1 < x_2 \in (-\infty, 1]$；
2. 因 $g(x)$ 在 $(-\infty, 1]$ 严格递增，故 $g(x_1) = -x_1^2 + 2x_1 < -x_2^2 + 2x_2 = g(x_2)$；
3. 因 $f(u) = e^u$ 在 $\mathbb{R}$ 上严格递增，且 $g(x_1) < g(x_2)$，故 $f(g(x_1)) < f(g(x_2))$，即 $e^{-x_1^2 + 2x_1} < e^{-x_2^2 + 2x_2}$；
4. 由 $x_1 < x_2$ 可推出 $h(x_1) < h(x_2)$，故 $h(x)$ 在 $(-\infty, 1]$ 上严格递增。

### 例题 3：CS/AI 场景应用题（二分查找与单调性）

**题目**：在神经网络超参数搜索中，学习率 $\eta$ 的搜索区间为 $(0.001, 0.1]$，已知损失函数 $L(\eta)$ 在该区间上严格递减（$\eta$ 越大，损失越小），请设计二分查找算法找到使 $L(\eta) < 0.01$ 的最小 $\eta$，并说明单调性在算法中的作用。

**解析**：

1. **单调性的核心作用**：严格递减意味着“$\eta$ 越大，$L(\eta)$ 越小”，因此一旦找到某个 $\eta_0$ 满足 $L(\eta_0) < 0.01$，所有大于 $\eta_0$ 的 $\eta$ 也满足条件，最小 $\eta$ 必在左半区间；反之，若 $L(\eta_0) \geq 0.01$，则最小 $\eta$ 必在右半区间，这是二分查找的前提。
2. **二分查找算法步骤**：
   - 初始化左边界 $low = 0.001$，右边界 $high = 0.1$；
   - 循环：计算中点 $\eta_{mid} = (low + high) / 2$，计算 $L(\eta_{mid})$；
     - 若 $L(\eta_{mid}) < 0.01$：更新 $high = \eta_{mid}$（最小 $\eta$ 在左半区间）；
     - 若 $L(\eta_{mid}) \geq 0.01$：更新 $low = \eta_{mid}$（最小 $\eta$ 在右半区间）；
   - 当 $high - low < 1e-6$（精度阈值），输出 $high$（最小满足条件的 $\eta$）。
3. **代码实现（简化版）**：

```python
def loss_function(eta):
    """模拟损失函数（严格递减）"""
    return 0.1 / eta  # 随 eta 增大而减小

def binary_search_min_eta(low, high, target=0.01, eps=1e-6):
    while high - low > eps:
        mid = (low + high) / 2
        l = loss_function(mid)
        if l < target:
            high = mid  # 满足条件，尝试更小的 eta
        else:
            low = mid  # 不满足条件，增大 eta
    return high

min_eta = binary_search_min_eta(0.001, 0.1)
print(f"最小满足条件的学习率：{min_eta:.6f}")
print(f"对应的损失值：{loss_function(min_eta):.6f}")
```


## 5. 典型应用场景（CS/AI 专项）

### 场景 1：二分查找与有序数据处理

- **核心依赖**：单调函数的“顺序可预测性”，确保每次查找可缩小一半搜索范围；
- **具体应用**：
  - 数据库索引：B+ 树索引的有序性（单调递增）支持二分查找，将查询时间复杂度从 $O(n)$ 降至 $O(\log n)$；
  - 模型参数搜索：在单调损失函数上二分查找最优超参数（如学习率、正则化系数），提升搜索效率。

### 场景 2：神经网络激活函数设计

- **核心依赖**：单调性保证梯度传播的稳定性，避免梯度震荡；
- **具体应用**：
  - 单调激活函数（ReLU、Leaky ReLU、GELU）：深层网络中梯度方向一致，减少训练波动，是 ResNet、Transformer 等模型的核心激活函数；
  - 非单调激活函数（Sigmoid、Tanh）：仅用于浅层网络或输出层，其非单调性易导致深层梯度消失。

### 场景 3：推荐系统与排序算法

- **核心依赖**：用户偏好得分函数的单调性，保证“更优特征对应更高推荐优先级”；
- **具体应用**：
  - 协同过滤：用户相似度得分函数单调递增（相似度越高，推荐权重越大）；
  - 排序模型（LambdaMART）：预测得分函数单调递增，确保排序结果符合用户偏好排序。

### 场景 4：优化算法与损失函数设计

- **核心依赖**：目标函数的单调性决定优化方向的确定性；
- **具体应用**：
  - 梯度下降：损失函数在当前迭代点附近的单调性，决定梯度反方向是最优下降方向；
  - 凸优化：凸函数的梯度是单调递增的，保证梯度下降收敛到全局最优解。

## 6. 经典证明题（函数章节专属，无超纲知识）

### 证明题 1：证明 $f(x) = \sqrt{x}$ 在 $[0, +\infty)$ 上严格递增

#### 已知

$f(x) = \sqrt{x}$，定义域 $[0, +\infty)$。

#### 求证

$f(x)$ 在 $[0, +\infty)$ 上严格递增。

#### 证明过程

1. 任取 $x_1 < x_2 \in [0, +\infty)$；

2. 计算 $f(x_2) - f(x_1) = \sqrt{x_2} - \sqrt{x_1}$，为消除根号，分子有理化：
   $$
   \sqrt{x_2} - \sqrt{x_1} = \frac{(\sqrt{x_2} - \sqrt{x_1})(\sqrt{x_2} + \sqrt{x_1})}{\sqrt{x_2} + \sqrt{x_1}} = \frac{x_2 - x_1}{\sqrt{x_2} + \sqrt{x_1}}
   $$

3. 分析符号：

   - 因 $x_1 < x_2$，故 $x_2 - x_1 > 0$；
   - 分母 $\sqrt{x_2} + \sqrt{x_1} \geq 0$，且仅当 $x_1 = x_2 = 0$ 时为 0，而 $x_1 < x_2$，故分母 $> 0$；

4. 因此 $\sqrt{x_2} - \sqrt{x_1} = \frac{x_2 - x_1}{\sqrt{x_2} + \sqrt{x_1}} > 0$，即 $f(x_2) > f(x_1)$；

5. 结论：$\forall x_1 < x_2 \in [0, +\infty)$，$f(x_1) < f(x_2)$，故 $f(x)$ 在 $[0, +\infty)$ 上严格递增。

### 证明题 2：证明“严格单调函数的反函数仍为严格单调函数”

#### 已知

$f(x)$ 在区间 $I$ 上严格递增，$J = f(I)$（$f$ 的值域），反函数 $f^{-1}(y)$ 满足 $y = f(x) \iff x = f^{-1}(y)$（$y \in J$）。

#### 求证

$f^{-1}(y)$ 在 $J$ 上严格递增。

#### 证明过程（反证法）

1. 任取 $y_1 < y_2 \in J$，需证明 $f^{-1}(y_1) < f^{-1}(y_2)$；
2. 由值域定义，存在 $x_1, x_2 \in I$ 使得 $y_1 = f(x_1)$，$y_2 = f(x_2)$，即 $x_1 = f^{-1}(y_1)$，$x_2 = f^{-1}(y_2)$；
3. 假设 $f^{-1}(y_1) \geq f^{-1}(y_2)$，即 $x_1 \geq x_2$；
4. 因 $f(x)$ 严格递增：
   - 若 $x_1 > x_2$，则 $f(x_1) > f(x_2)$，即 $y_1 > y_2$，与 $y_1 < y_2$ 矛盾；
   - 若 $x_1 = x_2$，则 $f(x_1) = f(x_2)$，即 $y_1 = y_2$，与 $y_1 < y_2$ 矛盾；
5. 假设不成立，故 $x_1 < x_2$，即 $f^{-1}(y_1) < f^{-1}(y_2)$；
6. 结论：$f^{-1}(y)$ 在 $J$ 上严格递增。

### 证明题 3：证明“若 $f(x)$ 在 $I$ 上严格递增，$g(x)$ 在 $I$ 上严格递减，则 $f(x) - g(x)$ 在 $I$ 上严格递增”

#### 已知

$f(x)$ 在 $I$ 上严格递增，$g(x)$ 在 $I$ 上严格递减。

#### 求证

$h(x) = f(x) - g(x)$ 在 $I$ 上严格递增。

#### 证明过程

1. 任取 $x_1 < x_2 \in I$；
2. 因 $f(x)$ 严格递增，故 $f(x_1) < f(x_2) \implies f(x_2) - f(x_1) > 0$；
3. 因 $g(x)$ 严格递减，故 $g(x_1) > g(x_2) \implies g(x_1) - g(x_2) > 0$；
4. 计算 $h(x_2) - h(x_1) = [f(x_2) - g(x_2)] - [f(x_1) - g(x_1)] = [f(x_2) - f(x_1)] + [g(x_1) - g(x_2)]$；
5. 两项均为正数，故 $h(x_2) - h(x_1) > 0 \implies h(x_2) > h(x_1)$；
6. 结论：$\forall x_1 < x_2 \in I$，$h(x_1) < h(x_2)$，故 $h(x)$ 在 $I$ 上严格递增。

## 7. 拓展与联系（知识体系定位）

### 7.2 后续延伸

- **微积分**：导数与单调性的关系（可导函数 $f'(x) > 0 \implies$ 严格递增，$f'(x) < 0 \implies$ 严格递减）；
- **实变函数**：单调函数的可积性（闭区间上单调函数必可积）与间断点性质（单调函数最多有可数个间断点）；
- **CS 算法**：排序算法（冒泡、快速排序的核心是构建单调序列）、贪心算法（部分贪心策略依赖目标函数单调性）；
- **AI 优化理论**：凸函数与单调性的等价关系（凸函数的梯度单调递增）、生成模型的单调可逆变换（Flow 模型、自回归模型）。
