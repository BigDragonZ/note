# 定义

##  有界函数

- **文字描述**：设函数 $f(x)$ 的定义域为 $D$，若存在一个正数 $M>0$，使得对**任意** $x \in D$（或 $x \in I \subseteq D$，即区间 $I$ 上有界），都有 $|f(x)| \leq M$，则称 $f(x)$ 在 $D$（或 $I$）上是有界函数。
- **数学描述**：
  $\exists M>0, \forall x \in D \, (\text{或 } I), \, |f(x)| \leq M$
- **等价定义**：若存在常数 $m$ 和 $M$（$m \leq M$），使得对任意 $x \in D$（或 $I$），都有 $m \leq f(x) \leq M$，则 $m$ 称为 $f(x)$ 的一个下界，$M$ 称为一个上界。

#### **有上界函数**

- **文字描述**：设函数 $f(x)$ 的定义域为 $D$，若存在常数 $M$（称为“上界”），使得对**任意** $x \in D$（或区间 $I \subseteq D$），都有 $f(x) \leq M$，则称 $f(x)$ 在 $D$（或 $I$）上是**有上界函数**。
- **数学描述**：$\exists M \in \mathbb{R}, \forall x \in D \, (\text{或 } I), f(x) \leq M$
- **示例**：$f(x) = -x^2$ 在 $\mathbb{R}$ 上有上界（取 $M=0$，对任意 $x$ 都有 $-x^2 \leq 0$）；$f(x) = \ln x$ 在 $(0, e]$ 上有上界（取 $M=1$，因 $\ln e=1$，且 $\ln x$ 在 $(0,e]$ 单调递增）。

#### **有下界函数**

- **文字描述**：设函数 $f(x)$ 的定义域为 $D$，若存在常数 $m$（称为“下界”），使得对**任意** $x \in D$（或区间 $I \subseteq D$），都有 $f(x) \geq m$，则称 $f(x)$ 在 $D$（或 $I$）上是**有下界函数**。
- **数学描述**：$\exists m \in \mathbb{R}, \forall x \in D \, (\text{或 } I), f(x) \geq m$
- **示例**：$f(x) = x^2$ 在 $\mathbb{R}$ 上有下界（取 $m=0$，对任意 $x$ 都有 $x^2 \geq 0$）；$f(x) = \frac{1}{x}$ 在 $(0, +\infty)$ 上有下界（取 $m=0$，因 $x>0$ 时 $\frac{1}{x}>0$）。

#### **单边有界与“有界函数”的关系**

- **核心定理**：函数 $f(x)$ 在区间 $I$ 上**有界** $\iff$ $f(x)$ 在 $I$ 上**既有上界又有下界**。
  - 证明（充分性）：若 $f(x)$ 有上界 $M$ 和下界 $m$，取 $M' = \max\{|M|, |m|\}$，则对任意 $x \in I$，有 $-M' \leq m \leq f(x) \leq M \leq M'$，即 $|f(x)| \leq M'$，故 $f(x)$ 有界；
  - 证明（必要性）：若 $f(x)$ 有界（存在 $M'>0$ 使 $|f(x)| \leq M'$），则 $-M' \leq f(x) \leq M'$，即 $f(x)$ 有上界 $M'$、下界 $-M'$。

##  无界函数

- **文字描述**：设函数 $f(x)$ 的定义域为 $D$，若对**任意**正数 $M>0$，都存在**某个** $x_0 \in D$（或 $x_0 \in I \subseteq D$），使得 $|f(x_0)| > M$，则称 $f(x)$ 在 $D$（或 $I$）上是无界函数。
- **数学描述**：
  $\forall M>0, \exists x_0 \in D \, (\text{或 } I), \, |f(x_0)| > M$



# 要点


#### 核心要素（有界函数的“三要素”）

1. **存在性（$\exists M>0$）**：“界”$M$ 只需存在即可，不唯一（若 $M=5$ 是界，则 $M=10$ 也是界）；
2. **普遍性（$\forall x \in D$ 或 $I$）**：函数值需对定义域/区间内的**所有** $x$ 满足 $|f(x)| \leq M$，而非部分 $x$；
3. **区间依赖性**：有界性是“相对概念”，需明确讨论的区间（同一函数在不同区间可能有界也可能无界，如 $f(x)=\frac{1}{x}$）

#### 核心原理（有界与无界的本质区别）

- **有界函数**：函数值的“波动范围被限制”，本质是“映射的输出集合是实数集的有界子集”（$f(D) \subseteq [-M, M]$），可理解为“输出不会超出固定边界”；
- **无界函数**：函数值的“波动范围无限制”，本质是“映射的输出集合是实数集的无界子集”（对任意 $M>0$，$f(D) \nsubseteq [-M, M]$），可理解为“存在某些输入使输出突破任意预设边界”。

## 关键公式（数学表达与推导）

##### （1）**有界函数的量化公式**

$$f(x) \text{ 在 } I \text{ 上有界} \iff \exists M>0, \forall x \in I, |f(x)| \leq M$$
**推导逻辑**：

- 充分性：若存在 $M$ 满足上述不等式，则 $f(x)$ 的上界为 $M$，下界为 $-M$，故有界；
- 必要性：若 $f(x)$ 有界，则存在上界 $M_1$ 和下界 $M_2$，取 $M=\max\{|M_1|, |M_2|\}$，则 $|f(x)| \leq M$。

##### （2）**无界函数的否定公式（反证法核心）**

$$f(x) \text{ 在 } I \text{ 上无界} \iff \forall M>0, \exists x_0 \in I, |f(x_0)| > M$$
**推导逻辑**：
无界是有界的否定，将有界定义中的“$\exists M$”改为“$\forall M$”，“$\forall x$”改为“$\exists x_0$”，“$\leq$”改为“$>$”，符合逻辑否定规则。



# 例题


## 1、基础定义法

证明 $f(x) = \frac{1}{x^2 + 1}$ 在 $\mathbb{R}$ 上有界

#### 已知

函数 $f(x) = \frac{1}{x^2 + 1}$，定义域为 $\mathbb{R}$（全体实数）。

#### 求证

$f(x)$ 在 $\mathbb{R}$ 上有界（即存在常数 $M>0$，对任意 $x \in \mathbb{R}$，都有 $|f(x)| \leq M$）。

#### 证明思路（定义法核心：找到一个固定的 $M$，满足定义）

1. 紧扣“有界函数定义”：需找到一个正数 $M$，使得对所有实数 $x$，$|f(x)| \leq M$；
2. 利用不等式性质分析 $f(x)$ 的取值范围：$x^2$ 是实数的平方，恒非负，由此推导 $x^2 + 1$ 的范围，再推导 $\frac{1}{x^2 + 1}$ 的范围；
3. 验证找到的 $M$ 满足定义。

#### 详细证明过程

1. **分析函数表达式的取值范围**：
   对任意实数 $x$，由平方的非负性可知 $x^2 \geq 0$；
   两边加 1，得 $x^2 + 1 \geq 0 + 1 = 1$（不等式两边同时加正数，不等号方向不变）；
   因 $x^2 + 1 \geq 1$，且 $x^2 + 1 > 0$（分母恒正），两边取倒数（正数取倒数，不等号方向反转），得：
   $$0 < \frac{1}{x^2 + 1} \leq 1$$

2. **验证有界函数定义**：
   由上述不等式可知，对任意 $x \in \mathbb{R}$，$|f(x)| = \left| \frac{1}{x^2 + 1} \right| = \frac{1}{x^2 + 1} \leq 1$；
   取 $M = 1$（$M$ 是固定正数），则满足“存在 $M>0$，对任意 $x \in \mathbb{R}$，$|f(x)| \leq M$”。

3. **结论**：
   故 $f(x) = \frac{1}{x^2 + 1}$ 在 $\mathbb{R}$ 上有界。

#### 思路总结（可复制）

证明“有界”→ 找固定 $M$ → 利用函数表达式的代数性质（平方、不等式）推导取值范围 → 验证 $M$ 满足定义。

## 2、单边有界证明

#### 已知

函数 $f(x) = x^2$，区间 $I = (0, +\infty)$（正实数集）。

#### 求证

1. $f(x)$ 在 $(0, +\infty)$ 上有下界；
2. $f(x)$ 在 $(0, +\infty)$ 上无上界。

#### 证明思路

1. 有下界：找固定常数 $m$，对任意 $x \in (0, +\infty)$，$f(x) \geq m$（利用平方恒正性）；
2. 无上界：紧扣“无上界定义”→ 对任意给定的正数 $M$，总能找到 $x_0 \in (0, +\infty)$，使得 $f(x_0) > M$（构造具体的 $x_0$ 满足条件）。

#### 详细证明过程

##### （1）证明 $f(x)$ 在 $(0, +\infty)$ 上有下界

- 由平方的性质：对任意 $x \in (0, +\infty)$，$x$ 是正数，故 $x^2 = x \cdot x > 0$；
- 取 $m = 0$（固定常数），则对任意 $x \in (0, +\infty)$，都有 $f(x) = x^2 \geq 0 = m$；
- 满足“有下界函数定义”（存在 $m$，对任意 $x \in I$，$f(x) \geq m$），故 $f(x)$ 在 $(0, +\infty)$ 上有下界。

##### （2）证明 $f(x)$ 在 $(0, +\infty)$ 上无上界

- 无上界定义：对**任意** $M>0$，**存在** $x_0 \in (0, +\infty)$，使得 $f(x_0) > M$；
- 对任意给定的正数 $M$，构造 $x_0 = \sqrt{M + 1}$（验证 $x_0 \in (0, +\infty)$）：
  因 $M>0$，故 $M + 1 > 1 > 0$，则 $\sqrt{M + 1} > 0$，即 $x_0 \in (0, +\infty)$；
- 计算 $f(x_0)$：$f(x_0) = x_0^2 = (\sqrt{M + 1})^2 = M + 1$；
- 因 $M + 1 > M$（任意正数加 1 大于自身），故 $f(x_0) = M + 1 > M$；
- 由 $M$ 的任意性（$M$ 可以是任意大的正数），不存在固定的上界 $M$ 满足“对所有 $x \in (0, +\infty)$，$f(x) \leq M$”，故 $f(x)$ 在 $(0, +\infty)$ 上无上界。

#### 思路总结（可复制）

- 证明“有下界”→ 找固定 $m$ → 利用函数代数性质（平方、绝对值）推导 $f(x) \geq m$；
- 证明“无上界”→ 对任意 $M$ → 构造 $x_0$（通常是关于 $M$ 的表达式）→ 验证 $f(x_0) > M$。

## 3、性质证明，逻辑推导

证明“两个有界函数的和仍为有界函数”

#### 已知

函数 $f(x)$ 和 $g(x)$ 在区间 $I$ 上均有界。

#### 求证

$h(x) = f(x) + g(x)$ 在区间 $I$ 上有界。

#### 证明思路（利用已知条件→推导新函数的界）

1. 由 $f(x)$、$g(x)$ 有界，写出它们的定义（找到各自的界 $M_1$、$M_2$）；
2. 利用绝对值不等式 $|a + b| \leq |a| + |b|$，推导 $|h(x)| = |f(x) + g(x)|$ 的范围；
3. 找到 $h(x)$ 的界 $M = M_1 + M_2$，验证满足定义。

#### 详细证明过程

1. **写出 $f(x)$、$g(x)$ 的有界定义**：
   因 $f(x)$ 在 $I$ 上有界，故存在 $M_1 > 0$，对任意 $x \in I$，$|f(x)| \leq M_1$；
   因 $g(x)$ 在 $I$ 上有界，故存在 $M_2 > 0$，对任意 $x \in I$，$|g(x)| \leq M_2$。

2. **推导 $h(x) = f(x) + g(x)$ 的范围**：
   对任意 $x \in I$，由绝对值不等式 $|a + b| \leq |a| + |b|$，得：
   $$|h(x)| = |f(x) + g(x)| \leq |f(x)| + |g(x)|$$
   代入 $|f(x)| \leq M_1$ 和 $|g(x)| \leq M_2$，得：
   $$|h(x)| \leq M_1 + M_2$$

3. **验证有界定义**：
   取 $M = M_1 + M_2$（因 $M_1$、$M_2$ 是正数，故 $M>0$），则对任意 $x \in I$，$|h(x)| \leq M$，满足“有界函数定义”。

4. **结论**：
   故 $h(x) = f(x) + g(x)$ 在区间 $I$ 上有界。

#### 思路总结（可复制）

证明“函数运算后的有界性”→ 利用已知函数的界 → 借助基本不等式（绝对值、平方等）→ 推导新函数的界 → 验证定义。

## 4、对比证明，强化定义理解

证明 $f(x) = \frac{1}{x}$ 在 $[1, +\infty)$ 上有界，在 $(0,1)$ 上无界

#### 已知

函数 $f(x) = \frac{1}{x}$，区间 $I_1 = [1, +\infty)$，$I_2 = (0,1)$。

#### 求证

1. $f(x)$ 在 $[1, +\infty)$ 上有界；
2. $f(x)$ 在 $(0,1)$ 上无界。

#### 证明思路

1. 有界性：利用区间 $[1, +\infty)$ 中 $x \geq 1$ 的性质，推导 $\frac{1}{x}$ 的范围；
2. 无界性：对任意 $M>0$，构造 $x_0 \in (0,1)$ 使得 $\frac{1}{x_0} > M$。

#### 详细证明过程

##### （1）证明 $f(x)$ 在 $[1, +\infty)$ 上有界

- 对任意 $x \in [1, +\infty)$，有 $x \geq 1$（区间定义）；
- 因 $x \geq 1$ 且 $x>0$，取倒数得 $0 < \frac{1}{x} \leq 1$（正数取倒数，不等号反转）；
- 取 $M = 1$，则对任意 $x \in [1, +\infty)$，$|f(x)| = \frac{1}{x} \leq 1 = M$，满足有界定义；
- 故 $f(x)$ 在 $[1, +\infty)$ 上有界。

##### （2）证明 $f(x)$ 在 $(0,1)$ 上无界

- 无界定义：对**任意** $M>0$，**存在** $x_0 \in (0,1)$，使得 $|f(x_0)| > M$；
- 对任意给定的 $M>0$，构造 $x_0 = \frac{1}{M + 1}$（验证 $x_0 \in (0,1)$）：
  因 $M>0$，故 $M + 1 > 1$，则 $0 < \frac{1}{M + 1} < 1$，即 $x_0 \in (0,1)$；
- 计算 $f(x_0)$：$|f(x_0)| = \frac{1}{x_0} = M + 1 > M$；
- 由 $M$ 的任意性，不存在固定的 $M>0$ 满足“对所有 $x \in (0,1)$，$|f(x)| \leq M$”，故 $f(x)$ 在 $(0,1)$ 上无界。

#### 思路总结

同一函数在不同区间的有界性可能不同→ 证明时需紧扣“区间性质”：

- 有界：利用区间端点限制，推导函数取值范围；
- 无界：利用区间“无下限/无上限”的特点，构造突破任意 $M$ 的 $x_0$。

## 总结

### 8.1 证明“有界函数”的通用步骤

1. 明确目标：找到一个固定的正数 $M$；
2. 分析函数表达式：利用平方、绝对值、区间范围等性质，推导 $|f(x)|$ 的上界；
3. 验证 $M$：确保对定义域/区间内的**所有** $x$，都有 $|f(x)| \leq M$；
4. 下结论：满足定义，故函数有界。

### 8.2 证明“无界函数”的通用步骤

1. 明确目标：对**任意**给定的正数 $M$，找到一个 $x_0$（属于定义域/区间）；
2. 构造 $x_0$：根据函数表达式，构造一个与 $M$ 相关的 $x_0$（通常是 $M$ 的倒数、平方根等）；
3. 验证 $x_0$：确保 $x_0$ 属于定义域/区间，且 $|f(x_0)| > M$；
4. 下结论：由 $M$ 的任意性，满足无界定义，故函数无界。

### 8.3 关键技巧

- 常用不等式：$x^2 \geq 0$（平方非负）、$|a + b| \leq |a| + |b|$（绝对值三角不等式）、正数取倒数不等号反转；
- 构造 $x_0$ 的技巧：证明无界时，$x_0$ 通常是“能让函数值变大的点”（如 $f(x)=\frac{1}{x}$ 在 $(0,1)$ 中，$x_0$ 越接近 0，函数值越大）。
